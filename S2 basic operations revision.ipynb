**loc**

- **iloc** takes only numeric indexes **loc** takes numeric and label indexes

- Syntax: visa_df [row_index,column_names]

```python
visa_df.loc[[10,11,12],['case_status']]
```

```python
visa_df.loc[[10,11,12],[1,2,5]] # NW it is expecting label in column parameter
visa_df.loc[[10,20,30]] # Works 
visa_df.loc[[10],['continent','case_status']] # works
```

```python
visa_df['continent']  # W series will come
visa_df[['continent']] # W DF will come
visa_df[['continent','case_status']] # works
visa_df[1:5,'continent'] # NW 
visa_df.loc[1:5,['continent']] # Works
visa_df.loc[range(1,5),['continent']] # Works
visa_df.loc[[10,20,30],range(1,4)] # NW
visa_df.loc[[10,11],['continent','education_of_employee']]
visa_df.loc[:,'continent'] # series will come
visa_df.loc[:,['continent']] # works
```

- loc will work if we give rows as numeric value and columns as lables

```python
roes=range(10,20)
columns=['continent','full_time_position','case_status']
visa_df.loc[roes,columns]
```
